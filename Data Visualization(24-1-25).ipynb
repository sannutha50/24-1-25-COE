```python
import matplotlib.pyplot as plt

#LINE PLOT

x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.title(" Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.title(" Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

#adding colors
plt.plot(x,y,color='blue',linestyle='solid',linewidth=2,marker='o')
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[1000,1500,2000,2500,3000,3500]
plt.title("Monthly Savings")
plt.xlabel("Months (1-6)")
plt.ylabel("Savings Amount")
plt.plot(x,y,color='green',linestyle='dashed',marker='o')
```

```python
#BAR CHART
categories=['A','B','C','D','E']
values=[10,15,7,12,9]
plt.bar(categories,values,color=['skyblue','pink','yellow','purple','green'],edgecolor='black',width=0.5)
plt.title("BAR CHART")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()
```

```python
#HISTOGRAM
#bins:defines the intervals for  grouping the data
ages=[20,22,22,23,25,27,27,28,30,31,32,34,35,35,44,44,45,50,55,60,65,70]
b=[20,30,40,50,60,70]
plt.hist(ages,bins=b,color='skyblue',edgecolor='black')
plt.title("Age distribution of customers")
plt.xlabel("Age Ranges")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
#PIE PLOT/CHART
x=['Python','Ruby','C++','Java']
y=[40,10,20,30]
color=['yellow','pink','lightgreen','skyblue']
plt.pie(y,colors=color,autopct='%1.1f%%',labels=x,startangle=90)
plt.legend(loc="upper left")
plt.title("Pie chart")
plt.show()
```

```python
#SCATTER PLOT
x=[5,10,15,20,25]
y=[10,20,25,30,35]
plt.scatter(x,y,color='green',s=100,edgecolor='black')
plt.title("Scatter plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```
