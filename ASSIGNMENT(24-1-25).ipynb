```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('store_sales.csv')
df.set_index('store_id', inplace=True)
monthly_sales = df.drop(columns=['store_id', 'city'], errors='ignore').sum()  
monthly_sales.plot(color='pink',marker='o', figsize=(10, 6))
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales for Each Store')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('store_sales.csv')
data['Total_Order_Quantity'] = data.iloc[:, 2:14].sum(axis=1)
plt.figure(figsize=(10, 6))
plt.hist(data['Total_Order_Quantity'], bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Total Order Quantity', fontsize=16)
plt.xlabel('Total Order Quantity', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('KPMG_Customer_Demography.csv')

plt.figure(figsize=(8, 5))
plt.hist(df["AGE"], bins=10, color='lightcoral', edgecolor='black', alpha=0.7)
plt.title("Age Distribution of Customers", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('KPMG_Customer_Demography.csv')



plt.figure(figsize=(8, 5))
plt.scatter(df["AGE"], df["past_3_years_bike_related_purchases"], color='lightgreen', alpha=0.7, edgecolor='black')
plt.title("Scatter Plot: Bike Purchases vs Age", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Past 3 Years Bike Related Purchases", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('marksheet.csv')
plt.figure(figsize=(12, 8))
plt.boxplot([df['Science'], df['English'], df['History'], df['Maths']],vert=False, patch_artist=True)
plt.title('Boxplot for Marks Across Subjects')
plt.xlabel('Marks')
plt.yticks([1, 2, 3, 4], ['Science', 'English', 'History', 'Maths'])


plt.show()
```
